```python
import pymongo
import pandas as pd
```

```python
mongo = pymongo.MongoClient('mongodb://localhost:27017')
db = mongo['sample_airbnb']
collection = db['listingsAndReviews']
```

```python
Details = []
for i in collection.find():
    data = dict(Id=i['_id'],
                Listing_url=i.get('listing_url'),
                Name=i.get('name'),
                Description=i.get('description'),
                House_rules=i.get('house_rules'),
                Property_type=i.get('property_type'),
                Room_type=i.get('room_type'),
                Bed_type=i.get('bed_type'),
                Min_nights=int(i.get('minimum_nights', 0)),
                Max_nights=int(i.get('maximum_nights', 0)),
                Cancellation_policy=i.get('cancellation_policy'),
                Accomodates=i.get('accommodates'),
                Total_bedrooms=i.get('bedrooms'),
                Total_beds=i.get('beds'),
                Availability_365=i.get('availability', {}).get('availability_365'),
                Price=i.get('price'),
                Security_deposit=i.get('security_deposit'),
                Cleaning_fee=i.get('cleaning_fee'),
                Extra_people=i.get('extra_people'),
                Guests_included=i.get('guests_included'),
                No_of_reviews=i.get('number_of_reviews'),
                Review_scores=i.get('review_scores', {}).get('review_scores_rating'),
                Amenities=', '.join(i.get('amenities', [])),
                Host_id=i.get('host', {}).get('host_id'),
                Host_name=i.get('host', {}).get('host_name'),
                Street=i.get('address', {}).get('street'),
                Country=i.get('address', {}).get('country'),
                Country_code=i.get('address', {}).get('country_code'),
                Location_type=i.get('address', {}).get('location', {}).get('type'),
                Longitude=i.get('address', {}).get('location', {}).get('coordinates', [])[0],
                Latitude=i.get('address', {}).get('location', {}).get('coordinates', [])[1],
                Is_location_exact=i.get('address', {}).get('location', {}).get('is_location_exact')
                )
    Details.append(data)
```

```python
data_frame = pd.DataFrame(Details)
data_frame
```

```python
# checking Data types
data_frame.dtypes
```

```python
#Checking Null Values
data_frame.isna().sum()
```

```python
#Filling Null Values

# Replace missing values in 'Name' column with a placeholder value
data_frame['Name'].fillna('Not Mentioned', inplace=True)

# Replace missing values in 'Description' column with an empty string
data_frame['Description'].fillna('No Desciption Provided', inplace=True)

# Replace missing values in 'Total_bedrooms' and 'Total_beds' columns with their respective means
data_frame['Total_bedrooms'].fillna(data_frame['Total_bedrooms'].mean(), inplace=True)
data_frame['Total_beds'].fillna(data_frame['Total_beds'].mean(), inplace=True)

# Replace missing values in 'Security_deposit' and 'Cleaning_fee' columns with 0
data_frame['Security_deposit'].fillna(0, inplace=True)
data_frame['Cleaning_fee'].fillna(0, inplace=True)

# Replace missing values in 'Review_scores' column with 0 (assuming missing values indicate no reviews)
data_frame['Review_scores'].fillna(0, inplace=True)

# For 'House_rules' column, replace missing values with a placeholder message
data_frame['House_rules'].fillna('No specific house rules', inplace=True)

```
